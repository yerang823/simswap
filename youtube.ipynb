```python
url = 'https://youtu.be/0DN4WgDRCj4'

start = '00:01:38'
end = '00:01:57'

from pytube import YouTube
DOWNLOAD_FOLDER = "./downloads"
yt = YouTube(url)
stream = yt.streams.get_highest_resolution()
#stream = yt.streams.filter(progressive=True).last()

stream.download(DOWNLOAD_FOLDER, filename='temp.mp4')

!rm ./downloads/output.mp4

!ffmpeg -i ./downloads/temp.mp4 -ss $start -to $end ./downloads/output.mp4
!rm ./downloads/temp.mp4
```
